```python
import cv2
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
img=cv2.imread(r'C:\Users\sushmitha\Downloads\picture (2).jpeg')
plt.imshow(img)
plt.title('sush')
plt.show
```

```python
def show_rgb_image(img, title=None, conversion=cv2.COLOR_BGR2RGB):
    img = cv2.cvtColor(img, conversion)
    plt.imshow(img)
    if title is not None:
        plt.title(title)
        plt.imshow(img)
```

```python
show_rgb_image(img, 'sush')
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("rgb_image", gray_image)
img_invert = cv2.bitwise_not(gray_image)
```

```python
inverted_image = 255 - img
cv2.imshow("Inverted", inverted_image)
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('josh',size=10)
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('sketch',size=18)
rgb_sketch=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(img, inverted_blurred, scale=256.0)
cv2.imshow("sketch", pencil_sketch)
plt.imshow(pencil_sketch)
```
